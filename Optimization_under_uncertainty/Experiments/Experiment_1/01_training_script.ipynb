```python
import os
from matplotlib import pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
import torchmetrics

from resnet_utils import *
```

```python
# input
b_size   = 512
n_epochs = 100

# transforms
TraFo = transforms.Compose([
    transforms.PILToTensor(),
    #transforms.CenterCrop((28,28)),
    transforms.RandomRotation(0.1)
])

# dsets
dset_train = datasets.MNIST(root='/eagle/projects/candle_aesp/siebenschuh/PT', train=True,  transform=TraFo)
dset_test  = datasets.MNIST(root='/eagle/projects/candle_aesp/siebenschuh/PT', train=False, transform=TraFo)

# dset loader
loader_train = DataLoader(dataset=dset_train, batch_size=b_size, num_workers=5, shuffle=True)
loader_test  = DataLoader(dataset=dset_test, batch_size=b_size, num_workers=5, shuffle=True)

# device
device = torch.device('cuda:0')

# model
net = ResNet(depth=3, inputDim=(1,28,28))
net = net.to(device)

# optimizer
optimizer = optim.Adam(params=net.parameters(), lr=0.0000005, betas=(0.9, 0.999), weight_decay=0.005)
criterion = nn.CrossEntropyLoss()

# metric
accuracy = torchmetrics.Accuracy(task='multiclass', num_classes=10)
accuracy = accuracy.to(device)
```

```python
# training
# accuracy
acc = torchmetrics.Accuracy(task='multiclass', num_classes=10)

# 
acc_local    = 0.0
running_loss = 0.0
loss_list    = []
train_acc, test_acc = [], []

# 1st loop: epochs
for i in range(n_epochs):
    print(f'\nEpoch: {i}')
    
    # 2nd loop: train batches
    for j, batch in enumerate(loader_train):
        # unpack
        inputs, labels = batch
        # float
        inputs = inputs.type(torch.float)
        
        # -> device 13s
        inputs = inputs.to(device)
        labels = labels.to(device)
        
        # zero grad
        optimizer.zero_grad()
        
        # backprop
        outputs = net(inputs)                  # forward
        loss    = criterion(outputs, labels)   # loss
        loss.backward()                        # backward 
        optimizer.step()
        
        # loss
        running_loss += loss.item()
        loss_list.append(loss.item())
        
        # print
        if(j%b_size==b_size-1):
            print(f'Loss : {loss:.2f}')
            
    # track
    train_acc.append(acc(outputs.argmax(dim=1).cpu(), labels.cpu()))
    
    # store model
    if(i in [0, 5, 20, n_epochs-1]):
        path_loc = f'./checkpoints/resnet_{net.depth}_mnist_e_{i}'
        torch.save(net.state_dict(), path_loc)
    
    
    # validation
    with torch.set_grad_enabled(False):
        test_acc_loc = []
        for batch in loader_test:
            inputs, labels = batch
            
            # float
            inputs = inputs.type(torch.float)
            
            # -> device 13s
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            # predict
            preds = net(inputs).argmax(dim=1)
    
            # accuracy
            test_acc_loc.append(acc(preds.cpu(), labels.cpu()))
        # append
        test_acc.append(np.mean(test_acc_loc))

print('\nFinished\n')
```

```python
plt.plot(train_acc)
plt.plot(test_acc)
```

```python
# Plot the train_acc and test_acc lines
plt.plot(train_acc)
plt.plot(test_acc)
plt.title('ResNet-style Model (d=3) Checkpoints')
# Define the x-axis locations and their corresponding indices
x_indices = [0, 5, 20, 99]

# Iterate over the x-axis locations and their indices
for x_idx in x_indices:
    # Get the corresponding value from test_acc
    y_val = test_acc[x_idx]
    
    # Plot a red dot at the specified x and y coordinates
    plt.plot(x_idx, y_val, 'o', color='red')
    
    # Add text annotation next to the red dot
    plt.text(x_idx+(3 if x_idx!=20 else -10), y_val, f'{y_val:.2f}', va='center', ha='left', color='red', fontsize=12)
    

# Display the plot
plt.show()
```
