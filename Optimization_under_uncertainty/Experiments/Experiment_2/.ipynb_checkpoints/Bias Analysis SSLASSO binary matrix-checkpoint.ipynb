```python
import numpy as np
from matplotlib import pyplot as plt

from SSLASSO import SSLASSO
from utils import expand, get_alpha
```

```python
# INPUTS
n = 1500
d = 20
k = 2
sigma_eps = 2.1

np.random.seed(590)
X_raw      = np.random.binomial(n=1, p=0.5, size=n*d).reshape(n,-1)
#X_raw       =  np.random.normal(size=n*d).reshape(n,-1)
X          = expand(X_raw, k=k, intercept=True)
alpha      = get_alpha(d=d, k=k, p_zero=0.7)

# sample response
y = np.random.normal(X @ alpha, scale=sigma_eps)
```

```python
ss = SSLASSO()

# set data
ss.setXy(X=X_raw, y=y)

# get estimate
ss.fit(method='MAP')

```

```python
# get estimate
alpha_hat = ss.get_alphas()
```

```python
plt.plot(alpha, 'o', markersize=9)  # Increase markersize to your preference

# Create a scatter plot for 'alpha_hat' with red markers
plt.plot(alpha_hat, 'o', c='red', markersize=6, alpha=0.75)  # Increase markersize to your preference

plt.show()
```

## Bias Analysis

## (!) Bias (over-shrinkage) is definitely an artifact from standardization of X but failing to back translate the coefficient vector alpha back to original units
- Even in the absence of noise `sigma_eps=0` and an exorbitant amount of data `n=50,000` as compared to the moderate dimension size (`d=8` -> p=`37`), the MAP coefficiient estimates of `alpha` are biased; in particular **overshrunk**
- overshrinkage seems *not* related to `sparsity` of the vector at hand and not too much to noise standard deviation (although the higher the `sigma`, the more likely a coefficient is shrunken to zero)
- For low `n`

- Compare, for example, the same `alpha` for `d=8`, `k=2` (--> `p=37`) and `sigma_eps=1.1`, `n=2000`, `p_zero=0.7`. Practically all coefficients overshrunk for binary matrix (top figure) while mostly unbiased (even under-shrunk) for the real-valued matrix.
